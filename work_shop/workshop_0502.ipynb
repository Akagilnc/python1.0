```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
import pandas as pd

def get_single_page_data(input_browser):
    titles = input_browser.find_elements_by_css_selector('#resultList > div')
    infos_list = []
    for infos in titles:
        if infos.get_attribute('class') != 'el':
            continue
        job_infos = infos.find_elements_by_tag_name('span')

        infos_list.append({
            '职位名': job_infos[0].text,
            '职位链接': job_infos[0].find_element_by_tag_name('a').get_attribute('href'),
            '公司名': job_infos[1].text,
            '工作地点': job_infos[2].text,
            '薪资': job_infos[3].text,
            '发布时间': job_infos[4].text 
        })
        
    return pd.DataFrame(infos_list)
```
